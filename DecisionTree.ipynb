```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix , roc_auc_score, roc_curve
from pandas_profiling import ProfileReport
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train , x_test, y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state=300)
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_model.score(x_test, y_test)
```

```python
path = dt_model1.cost_complexity_pruning_path(x1,y1)
ccp_alpha = path['ccp_alphas']
```

```python
path
# ccp is cost complexity of pruning
```

```python
dt_model2 = []
for ccp in ccp_alpha:
    dt_m = DecisionTreeClassifier(ccp_alpha=ccp)
    dt_m.fit(x1,y1)
    dt_model2.append(dt_m)
```

```python
dt_model2
```

```python
train_score = [i.score(x1,y1) for i in dt_model2]
```

```python
test_score = [i.score(x_test, y_test ) for i in dt_model2]
```

```python
# train and test score with resp to ccp alpha
```

```python
fig,ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel('accuracy')
ax.set_title('Accuracy vs alpha for training and testing sets')
ax.plot(ccp_alpha, train_score, marker = 'o', label= 'train', drawstyle = 'steps-post')
ax.plot(ccp_alpha, test_score, marker = 'o', label = 'test', drawstyle = 'steps-post')
ax.legend()
plt.show()
```
